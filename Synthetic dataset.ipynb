```python
import pyaudio
import wave

# Set parameters for audio recording
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
CHUNK = 1024
RECORD_SECONDS = 5
OUTPUT_FILENAME = "zaid.wav"

# Initialize PyAudio
audio = pyaudio.PyAudio()

# Open stream for audio recording
stream = audio.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)

print("Recording...")

frames = []

# Record audio for specified duration
for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK)
    frames.append(data)

print("Finished recording.")

# Stop and close the stream
stream.stop_stream()
stream.close()
audio.terminate()

# Write audio data to WAV file
with wave.open(OUTPUT_FILENAME, 'wb') as wf:
    wf.setnchannels(CHANNELS)
    wf.setsampwidth(audio.get_sample_size(FORMAT))
    wf.setframerate(RATE)
    wf.writeframes(b''.join(frames))

print("Audio saved as:", OUTPUT_FILENAME)


```

```python
from typing import List
from pydantic import BaseModel, Field, constr, condecimal
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field
class Credit(BaseModel):
    Age: int =Field(description="Age of the customer")
    Occupation: str =Field(description="Histroy of the payment he/she has done")
    Income: int  =Field(description="How much does customer earn yearly  in USD ($)")
    Credit_Score: int =Field(description="numerical representation of an individual's creditworthiness, indicating their likelihood of repaying borrowed money.")   
    Credit_History:str =Field(description=" refers to a record of an individual's borrowing and repayment activities over time. It includes information about past loans, credit cards, and other forms of credit, as well as payment history, balances owed, and any instances of late or missed payments")
    Loan_Type: str=Field(description="Refers to the specific category or purpose of the loan that an individual applies for or obtains from a lender.")
    Loan_Amount: int=Field(description="Amount of money customer took from the bank")
    Interest_Rate: int=Field(description="Intrest bank has charged in percent (%)")
    Loan_Term: int=Field(description="Time taken from bank to repay in years")
    Monthly_Payment: float=Field(description="Monthly payment customer is obliged  to pay in USD $")
    Total_Interest_Paid: float=Field(description="Total amount of intreset customer has payed")
    Total_Repayment_Amount: float=Field(description="Amount to be paid by the custmoer")
        
```

```python
from langchain_cohere import ChatCohere
from langchain_groq import ChatGroq
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import ChatMessagePromptTemplate, PromptTemplate
from langchain.prompts import ChatPromptTemplate
def profile(df):
    llm = ChatGroq(model="llama3-8b-8192", temperature=0)
    parser = JsonOutputParser(pydantic_object=Credit)  
    prompt = PromptTemplate(
        template="""Here is the name of the person  {name},their credit history {credit_history},thier  Payment_history \n {Payment_history}, Financial_circumstances \n {Financial_circumstances},\n{Communication}
        Looking at this things Make me fake/synthitic Age, Occupation ,Income, Credit_Score, Credit_History ,Loan_Type ,Loan_Amount, Interest_Rate ,Loan_Term ,Monthly_Payment
        Total_Interest_Paid ,Total_Repayment_Amount
        Make sure you put the income amount in USD and interest rate in %
         \n{format_instructions} 
        """,
        input_variables=["name","credit_history","Payment_history","Financial_circumstances","Communication"],
        partial_variables={"format_instructions": parser.get_format_instructions()})
    chain = prompt | llm | parser
    answer=chain.invoke({'name':df['Name'],"credit_history":df['Credit_worthiness'],"Payment_history":df['Payment_history'],"Financial_circumstances":df['Financial_circumstances'],"Communication":df['Communication']})
    return answer
```
